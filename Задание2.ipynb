```python
# 1. РқР°СҒСӮСҖРҫР№РәР° РІРҫСҒРҝСҖРҫРёР·РІРҫРҙРёРјРҫСҒСӮРё
seed_value = 0

import os
os.environ['PYTHONHASHSEED'] = str(seed_value)
os.environ['TF_CUDNN_DETERMINISTIC'] = '1'

import random
random.seed(seed_value)

import numpy as np
np.random.seed(seed_value)

import tensorflow as tf
tf.random.set_seed(seed_value)

# 2. РҳРјРҝРҫСҖСӮ РұРёРұР»РёРҫСӮРөРә
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout

sns.set_style('whitegrid')

# 3. Р—Р°РіСҖСғР·РәР° Рё РҫРұСҖР°РұРҫСӮРәР° РҙР°РҪРҪСӢС…
# Р—Р°РіСҖСғР¶Р°РөРј С„Р°Р№Р» (РҝСҖРөРҙРҝРҫР»Р°РіР°РөРј, СҮСӮРҫ РҫРҪ РІ СӮРөРәСғСүРөР№ РҙРёСҖРөРәСӮРҫСҖРёРё)
df_dirty = pd.read_csv('MSFT_corrupted.csv')

# РҹСҖРҫРІРөСҖРәР° РҪР° NaN
print("РҹСҖРҫРҝСғСүРөРҪРҪСӢРө Р·РҪР°СҮРөРҪРёСҸ РҙРҫ РҫРұСҖР°РұРҫСӮРәРё:")
print(df_dirty.isnull().sum())

# Р’РҫСҒСҒСӮР°РҪРҫРІР»РөРҪРёРө РҝСҖРҫРҝСғСүРөРҪРҪСӢС… Р·РҪР°СҮРөРҪРёР№ (Р»РёРҪРөР№РҪР°СҸ РёРҪСӮРөСҖРҝРҫР»СҸСҶРёСҸ)
df = df_dirty.interpolate(method='linear')

print("\nРҹСҖРҫРҝСғСүРөРҪРҪСӢРө Р·РҪР°СҮРөРҪРёСҸ РҝРҫСҒР»Рө РёРҪСӮРөСҖРҝРҫР»СҸСҶРёРё:")
print(df.isnull().sum())

# РҹРҫСҒСӮСҖРҫРөРҪРёРө РіСҖР°С„РёРәР° СҶРөРҪСӢ Р·Р°РәСҖСӢСӮРёСҸ
plt.figure(figsize=(16, 6))
plt.plot(df['date'], df['close'], label='РҰРөРҪР° Р·Р°РәСҖСӢСӮРёСҸ', color='blue')
plt.title('РҰРөРҪР° Р°РәСҶРёР№ Microsoft Р·Р° РІСҖРөРјСҸ')
plt.xlabel('Р”Р°СӮР°')
plt.ylabel('РҰРөРҪР° ($)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# РһСҒСӮР°РІР»СҸРөРј СӮРҫР»СҢРәРҫ РәРҫР»РҫРҪРәСғ close
train_df = df[['close']].copy()

# 4. РҹР°СҖР°РјРөСӮСҖСӢ РјРҫРҙРөР»Рё
EPOCHS = 50          # РәРҫР»РёСҮРөСҒСӮРІРҫ СҚРҝРҫС…
DAYS = 10            # РәРҫР»РёСҮРөСҒСӮРІРҫ РҙРҪРөР№ РҙР»СҸ РҫРәРҪР°
SPLIT_VALUE = 0.95   # РҙРҫР»СҸ СӮСҖРөРҪРёСҖРҫРІРҫСҮРҪСӢС… РҙР°РҪРҪСӢС…

# Р Р°Р·РҙРөР»РөРҪРёРө РҪР° СӮСҖРөРҪРёСҖРҫРІРҫСҮРҪСғСҺ Рё СӮРөСҒСӮРҫРІСғСҺ СҮР°СҒСӮРё
training_data_len = int(len(train_df) * SPLIT_VALUE)
print(f"\nР”Р»РёРҪР° СӮСҖРөРҪРёСҖРҫРІРҫСҮРҪСӢС… РҙР°РҪРҪСӢС…: {training_data_len}")
print(f"Р”Р»РёРҪР° СӮРөСҒСӮРҫРІСӢС… РҙР°РҪРҪСӢС…: {len(train_df) - training_data_len}")

# РңР°СҒСҲСӮР°РұРёСҖРҫРІР°РҪРёРө РҙР°РҪРҪСӢС…
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(train_df.values)

# 5. РЎРҫР·РҙР°РҪРёРө СӮСҖРөРҪРёСҖРҫРІРҫСҮРҪРҫРіРҫ РҪР°РұРҫСҖР°
train_data = scaled_data[:training_data_len, :]

x_train = []
y_train = []

for i in range(DAYS, len(train_data)):
    x_train.append(train_data[i-DAYS:i])
    y_train.append(train_data[i, 0])

x_train = np.array(x_train)
y_train = np.array(y_train)

print(f"РӨРҫСҖРјР° x_train: {x_train.shape}")
print(f"РӨРҫСҖРјР° y_train: {y_train.shape}")

# 6. РҹРҫСҒСӮСҖРҫРөРҪРёРө РјРҫРҙРөР»Рё LSTM
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(x_train.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(units=1))

model.compile(optimizer='adam', loss='mean_squared_error')
model.summary()

# 7. РһРұСғСҮРөРҪРёРө РјРҫРҙРөР»Рё
history = model.fit(
    x_train, y_train,
    batch_size=32,
    epochs=EPOCHS,
    validation_split=0.1,
    verbose=1
)

# 8. РЎРҫР·РҙР°РҪРёРө СӮРөСҒСӮРҫРІРҫРіРҫ РҪР°РұРҫСҖР°
test_data = scaled_data[training_data_len - DAYS:, :]

x_test = []
y_test = []

for i in range(DAYS, len(test_data)):
    x_test.append(test_data[i-DAYS:i])
    y_test.append(test_data[i, 0])

x_test = np.array(x_test)
y_test = np.array(y_test)

print(f"\nРӨРҫСҖРјР° x_test: {x_test.shape}")
print(f"РӨРҫСҖРјР° y_test: {y_test.shape}")

# 9. РҹРҫР»СғСҮРөРҪРёРө РҝСҖРөРҙСҒРәР°Р·Р°РҪРёР№
predictions = model.predict(x_test)

# РһРұСҖР°СӮРҪРҫРө РјР°СҒСҲСӮР°РұРёСҖРҫРІР°РҪРёРө
predictions_original = scaler.inverse_transform(predictions)
y_test_original = scaler.inverse_transform(y_test.reshape(-1, 1))

# РЎРҫР·РҙР°РҪРёРө DataFrame РҙР»СҸ РІРёР·СғР°Р»РёР·Р°СҶРёРё
train = train_df[:training_data_len]
valid = train_df[training_data_len:].copy()
valid['predictions'] = predictions_original

# 10. Р’РёР·СғР°Р»РёР·Р°СҶРёСҸ СҖРөР·СғР»СҢСӮР°СӮРҫРІ LSTM
plt.figure(figsize=(16, 6))
plt.plot(train['close'].values, label='РўСҖРөРҪРёСҖРҫРІРҫСҮРҪСӢРө РҙР°РҪРҪСӢРө', color='blue', alpha=0.7)
plt.plot(valid['close'].values, label='РӨР°РәСӮРёСҮРөСҒРәР°СҸ СҶРөРҪР°', color='green', linewidth=2)
plt.plot(valid['predictions'].values, label='РҹСҖРөРҙСҒРәР°Р·Р°РҪРёСҸ LSTM', color='red', linewidth=1.5)
plt.title('РҹСҖРөРҙСҒРәР°Р·Р°РҪРёРө СҶРөРҪСӢ Р°РәСҶРёР№ СҒ РҝРҫРјРҫСүСҢСҺ LSTM')
plt.xlabel('Р”РҪРё')
plt.ylabel('РҰРөРҪР° ($)')
plt.legend(loc='lower right')
plt.show()

# 11. Step-1 Baseline РјРҫРҙРөР»
# РЎРҫР·РҙР°РөРј РҝСҖРөРҙСҒРәР°Р·Р°РҪРёСҸ: РҝСҖРөРҙСҒРәР°Р·Р°РҪРёРө РҙР»СҸ РҙРҪСҸ i = Р·РҪР°СҮРөРҪРёРө РҙРҪСҸ i-1
predictions_step = y_test_original.copy()  # РҡРҫРҝРёСҖСғРөРј
predictions_step = np.roll(predictions_step, 1, axis=0)  # РЎРҙРІРёРіР°РөРј РІРҪРёР·
predictions_step[0] = predictions_step[1]  # Р—Р°РҝРҫР»РҪСҸРөРј РҝРөСҖРІРҫРө Р·РҪР°СҮРөРҪРёРө

valid['predictions_step'] = predictions_step

# РЎСҖР°РІРҪРөРҪРёРө LSTM Рё Step-1
plt.figure(figsize=(16, 6))
plt.plot(valid['close'].values, label='РӨР°РәСӮРёСҮРөСҒРәР°СҸ СҶРөРҪР°', linewidth=2, color='black')
plt.plot(valid['predictions'].values, label='LSTM', linewidth=1.5, color='blue')
plt.plot(valid['predictions_step'].values, label='Step-1 (РҝСҖРөРҙСӢРҙСғСүРёР№ РҙРөРҪСҢ)', linewidth=1.5, linestyle='--', color='red')
plt.title('РЎСҖР°РІРҪРөРҪРёРө LSTM Рё Step-1 РјРҫРҙРөР»РөР№')
plt.xlabel('Р”РҪРё')
plt.ylabel('РҰРөРҪР° ($)')
plt.legend(loc='lower right')
plt.show()

# Р’СӢСҮРёСҒР»РөРҪРёРө РјРөСӮСҖРёРә
rmse_lstm = np.sqrt(mean_squared_error(valid['close'].values, valid['predictions'].values))
rmse_step1 = np.sqrt(mean_squared_error(valid['close'].values[1:], valid['predictions_step'].values[1:]))

mae_lstm = mean_absolute_error(valid['close'].values, valid['predictions'].values)
mae_step1 = mean_absolute_error(valid['close'].values[1:], valid['predictions_step'].values[1:])

print("\n" + "="*50)
print("РЎР РҗР’РқР•РқРҳР• РңРһР”Р•РӣР•Рҷ")
print("="*50)
print(f"LSTM - RMSE: {rmse_lstm:.4f}, MAE: {mae_lstm:.4f}")
print(f"Step-1 - RMSE: {rmse_step1:.4f}, MAE: {mae_step1:.4f}")

# 12. РҗРІСӮРҫРҪРҫРјРҪРҫРө РҝСҖРөРҙСҒРәР°Р·Р°РҪРёРө (РёР· РҫРҙРҪРҫР№ СӮРҫСҮРәРё)
x_test_independent = test_data[:DAYS].copy()
predictions_independent = []

for i in range(len(valid)):
    x_input = x_test_independent[-DAYS:].reshape(1, DAYS, 1)
    result = model.predict(x_input, verbose=0)
    predictions_independent.append(result[0, 0])
    x_test_independent = np.vstack([x_test_independent, result])

predictions_independent_original = scaler.inverse_transform(np.array(predictions_independent).reshape(-1, 1))

valid['predictions_indep'] = predictions_independent_original

# Р’РёР·СғР°Р»РёР·Р°СҶРёСҸ Р°РІСӮРҫРҪРҫРјРҪСӢС… РҝСҖРөРҙСҒРәР°Р·Р°РҪРёР№
plt.figure(figsize=(16, 6))
plt.plot(valid['close'].values, label='РӨР°РәСӮРёСҮРөСҒРәР°СҸ СҶРөРҪР°', linewidth=2, color='black')
plt.plot(valid['predictions'].values, label='LSTM (СҒ СҖРөР°Р»СҢРҪСӢРјРё РҙР°РҪРҪСӢРјРё)', linewidth=1.5, color='blue')
plt.plot(valid['predictions_indep'].values, label='LSTM (Р°РІСӮРҫРҪРҫРјРҪСӢР№ РҝСҖРҫРіРҪРҫР·)', linewidth=1.5, linestyle='--', color='orange')
plt.title('РҗРІСӮРҫРҪРҫРјРҪРҫРө РҝСҖРөРҙСҒРәР°Р·Р°РҪРёРө LSTM')
plt.xlabel('Р”РҪРё')
plt.ylabel('РҰРөРҪР° ($)')
plt.legend(loc='lower right')
plt.show()

# 13. Р”РҫРұР°РІР»РөРҪРёРө СҒРәРҫР»СҢР·СҸСүРёС… СҒСҖРөРҙРҪРёС… (MA)
train_df_with_ma = df[['close']].copy()

# Р’СӢСҮРёСҒР»РөРҪРёРө MA Р·Р° 7, 14 Рё 28 РҙРҪРөР№
train_df_with_ma['MA_7'] = train_df_with_ma['close'].rolling(window=7).mean()
train_df_with_ma['MA_14'] = train_df_with_ma['close'].rolling(window=14).mean()
train_df_with_ma['MA_28'] = train_df_with_ma['close'].rolling(window=28).mean()

# Р—Р°РҝРҫР»РҪРөРҪРёРө РҝСҖРҫРҝСғСҒРәРҫРІ
train_df_with_ma = train_df_with_ma.fillna(method='bfill')

# Р’РёР·СғР°Р»РёР·Р°СҶРёСҸ СҒРәРҫР»СҢР·СҸСүРёС… СҒСҖРөРҙРҪРёС…
plt.figure(figsize=(16, 6))
plt.plot(train_df_with_ma['close'].values, label='РҰРөРҪР° Р·Р°РәСҖСӢСӮРёСҸ', alpha=0.5, color='blue')
plt.plot(train_df_with_ma['MA_7'].values, label='MA 7 РҙРҪРөР№', linewidth=1.5)
plt.plot(train_df_with_ma['MA_14'].values, label='MA 14 РҙРҪРөР№', linewidth=1.5)
plt.plot(train_df_with_ma['MA_28'].values, label='MA 28 РҙРҪРөР№', linewidth=1.5)
plt.title('РЎРәРҫР»СҢР·СҸСүРёРө СҒСҖРөРҙРҪРёРө РҙР»СҸ СҶРөРҪСӢ Р°РәСҶРёР№ Microsoft')
plt.xlabel('Р”РҪРё')
plt.ylabel('РҰРөРҪР° ($)')
plt.legend()
plt.show()

# 14. РңР°СҒСҲСӮР°РұРёСҖРҫРІР°РҪРёРө РҙР°РҪРҪСӢС… СҒ MA
scaler_ma = MinMaxScaler(feature_range=(0, 1))
scaled_data_ma = scaler_ma.fit_transform(train_df_with_ma.values)

# Р Р°Р·РҙРөР»РөРҪРёРө РҪР° СӮСҖРөРҪРёСҖРҫРІРҫСҮРҪСӢРө Рё СӮРөСҒСӮРҫРІСӢРө РҙР°РҪРҪСӢРө
training_data_len_ma = int(len(scaled_data_ma) * SPLIT_VALUE)
n_features = scaled_data_ma.shape[1]  # 4 РҝСҖРёР·РҪР°РәР°

# РЎРҫР·РҙР°РҪРёРө СӮСҖРөРҪРёСҖРҫРІРҫСҮРҪРҫРіРҫ РҪР°РұРҫСҖР° СҒ MA
train_data_ma = scaled_data_ma[:training_data_len_ma, :]

x_train_ma = []
y_train_ma = []

for i in range(DAYS, len(train_data_ma)):
    x_train_ma.append(train_data_ma[i-DAYS:i])
    y_train_ma.append(train_data_ma[i, 0])

x_train_ma = np.array(x_train_ma)
y_train_ma = np.array(y_train_ma)

print(f"\nРӨРҫСҖРјР° x_train_ma: {x_train_ma.shape}")

# 15. РҹРҫСҒСӮСҖРҫРөРҪРёРө РјРҫРҙРөР»Рё LSTM СҒ MA
model_ma = Sequential()
model_ma.add(LSTM(units=50, return_sequences=True, input_shape=(x_train_ma.shape[1], n_features)))
model_ma.add(Dropout(0.2))
model_ma.add(LSTM(units=50, return_sequences=False))
model_ma.add(Dropout(0.2))
model_ma.add(Dense(units=1))

model_ma.compile(optimizer='adam', loss='mean_squared_error')

# РһРұСғСҮРөРҪРёРө
history_ma = model_ma.fit(
    x_train_ma, y_train_ma,
    batch_size=32,
    epochs=EPOCHS,
    validation_split=0.1,
    verbose=1
)

# 16. РўРөСҒСӮРёСҖРҫРІР°РҪРёРө РјРҫРҙРөР»Рё СҒ MA
test_data_ma = scaled_data_ma[training_data_len_ma - DAYS:, :]

x_test_ma = []
y_test_ma = []

for i in range(DAYS, len(test_data_ma)):
    x_test_ma.append(test_data_ma[i-DAYS:i])
    y_test_ma.append(test_data_ma[i, 0])

x_test_ma = np.array(x_test_ma)
y_test_ma = np.array(y_test_ma)

# РҹРҫР»СғСҮРөРҪРёРө РҝСҖРөРҙСҒРәР°Р·Р°РҪРёР№
predictions_ma = model_ma.predict(x_test_ma)

# РһРұСҖР°СӮРҪРҫРө РјР°СҒСҲСӮР°РұРёСҖРҫРІР°РҪРёРө
# РЎРҫР·РҙР°РөРј РҝРҫР»РҪСӢР№ РјР°СҒСҒРёРІ РҙР»СҸ РҫРұСҖР°СӮРҪРҫРіРҫ РҝСҖРөРҫРұСҖР°Р·РҫРІР°РҪРёСҸ
temp_array = np.hstack([predictions_ma, np.zeros((len(predictions_ma), n_features - 1))])
predictions_ma_original = scaler_ma.inverse_transform(temp_array)[:, 0]

# Р”РҫРұР°РІР»РөРҪРёРө РІ DataFrame
valid_ma = train_df_with_ma[training_data_len_ma:].copy()
valid_ma['predictions_ma'] = predictions_ma_original

# 17. РЎСҖР°РІРҪРөРҪРёРө РІСҒРөС… РјРҫРҙРөР»РөР№
plt.figure(figsize=(16, 6))
plt.plot(valid_ma['close'].values, label='РӨР°РәСӮРёСҮРөСҒРәР°СҸ СҶРөРҪР°', linewidth=2, color='black')
plt.plot(valid['predictions'].values[:len(valid_ma)], label='LSTM (РұРөР· MA)', linewidth=1.5, color='blue')
plt.plot(valid_ma['predictions_ma'].values, label='LSTM + MA', linewidth=1.5, color='green')
plt.plot(valid['predictions_step'].values[:len(valid_ma)], label='Step-1', linewidth=1.5, linestyle='--', color='red')
plt.title('РЎСҖР°РІРҪРөРҪРёРө РІСҒРөС… РјРҫРҙРөР»РөР№')
plt.xlabel('Р”РҪРё')
plt.ylabel('РҰРөРҪР° ($)')
plt.legend(loc='lower right')
plt.show()

# 18. Р’СӢСҮРёСҒР»РөРҪРёРө РёСӮРҫРіРҫРІСӢС… РјРөСӮСҖРёРә
rmse_ma = np.sqrt(mean_squared_error(valid_ma['close'].values, valid_ma['predictions_ma'].values))
mae_ma = mean_absolute_error(valid_ma['close'].values, valid_ma['predictions_ma'].values)

print("\n" + "="*50)
print("РҳРўРһР“РһР’РһР• РЎР РҗР’РқР•РқРҳР• РңРһР”Р•РӣР•Рҷ")
print("="*50)
print(f"{'РңРҫРҙРөР»СҢ':<20} {'RMSE':<10} {'MAE':<10}")
print("-"*40)
print(f"{'LSTM (РұРөР· MA)':<20} {rmse_lstm:.4f}     {mae_lstm:.4f}")
print(f"{'LSTM + MA':<20} {rmse_ma:.4f}     {mae_ma:.4f}")
print(f"{'Step-1':<20} {rmse_step1:.4f}     {mae_step1:.4f}")

print("\n" + "="*50)
print("Р’Р«Р’РһР”Р«")
print("="*50)
print("1. РңРҫРҙРөР»СҢ Step-1 РҝРҫРәР°Р·Р°Р»Р° РҪР°РёР»СғСҮСҲРёРө СҖРөР·СғР»СҢСӮР°СӮСӢ, СҮСӮРҫ РҫРұСҠСҸСҒРҪСҸРөСӮСҒСҸ")
print("   РІСӢСҒРҫРәРҫР№ Р°РІСӮРҫРәРҫСҖСҖРөР»СҸСҶРёРөР№ СҶРөРҪ Р°РәСҶРёР№.")
print("2. Р”РҫРұР°РІР»РөРҪРёРө СҒРәРҫР»СҢР·СҸСүРёС… СҒСҖРөРҙРҪРёС… (MA) СғР»СғСҮСҲРёР»Рҫ РәР°СҮРөСҒСӮРІРҫ LSTM,")
print("   РҪРҫ РҪРө РҝСҖРөРІР·РҫСҲР»Рҫ РҝСҖРҫСҒСӮСғСҺ Step-1 РјРҫРҙРөР»СҢ.")
print("3. РҗРІСӮРҫРҪРҫРјРҪРҫРө РҝСҖРөРҙСҒРәР°Р·Р°РҪРёРө (РёР· РҫРҙРҪРҫР№ СӮРҫСҮРәРё) РҝСҖРёРІРөР»Рҫ Рә Р·РҪР°СҮРёСӮРөР»СҢРҪРҫРјСғ")
print("   СғС…СғРҙСҲРөРҪРёСҺ РәР°СҮРөСҒСӮРІР° РёР·-Р·Р° РҪР°РәРҫРҝР»РөРҪРёСҸ РҫСҲРёРұРҫРә.")
print("4. Р”Р»СҸ С„РёРҪР°РҪСҒРҫРІСӢС… РІСҖРөРјРөРҪРҪСӢС… СҖСҸРҙРҫРІ РҝСҖРҫСҒСӮСӢРө РјРҫРҙРөР»Рё СҮР°СҒСӮРҫ РҫРәР°Р·СӢРІР°СҺСӮСҒСҸ")
print("   РұРҫР»РөРө СҚС„С„РөРәСӮРёРІРҪСӢРјРё, СҮРөРј СҒР»РҫР¶РҪСӢРө РҪРөР№СҖРҫСҒРөСӮРөРІСӢРө Р°СҖС…РёСӮРөРәСӮСғСҖСӢ.")
```
